```python
%%sql

#Code 1
SELECT COUNT(DISTINCT(purchase_id)),
		COUNT(DISTINCT(product_id)),
		COUNT(DISTINCT(affiliate_id)),
		COUNT(DISTINCT(producer_id)),
		COUNT(DISTINCT(buyer_id)),
    COUNT(DISTINCT(product_category)),
    COUNT(DISTINCT(product_niche)),
    COUNT(DISTINCT(purchase_device)),
    COUNT(DISTINCT(purchase_origin))
FROM sales_data

#Code 2
SELECT SUM(is_origin_page_social_network) AS origin_network
FROM sales_data;

#Code 3
SELECT DISTINCT(product_category), 
		COUNT(DISTINCT(product_id)) AS count_dist, 
    COUNT(DISTINCT(product_id))/17883 AS dist_prop, 
    COUNT(purchase_id) AS count_sales, 
    COUNT(purchase_id)/1599828 AS sales_prop
FROM sales_data
GROUP BY product_category
ORDER BY count_sales DESC

#Code 4
SELECT DISTINCT(product_niche), 
		COUNT(DISTINCT(product_id)) AS count_dist, 
    COUNT(DISTINCT(product_id))/17883 AS dist_prop, 
    COUNT(product_niche) AS count_sales, 
    COUNT(product_niche)/1599828 AS sales_prop
FROM sales_data
GROUP BY product_niche
ORDER BY count_sales DESC

#Code 5
SELECT DISTINCT(purchase_device), 
    COUNT(purchase_device) AS count_sales, 
    COUNT(purchase_device)/1599828 AS sales_prop
FROM sales_data
GROUP BY purchase_device
ORDER BY count_sales DESC

#Code 6
SELECT DISTINCT(product_category),
		COUNT(purchase_id),
    COUNT(purchase_id)/664197,
    COUNT(purchase_id)/1599828
FROM sales_data
WHERE purchase_device = 'eReaders'
GROUP BY product_category

#Code 7
SELECT producer_id, 
    COUNT(producer_id) AS sales_count, 
    COUNT(producer_id)/1599828 AS sales_proportion
	FROM sales_data
	GROUP BY producer_id
	ORDER BY sales_count DESC
    LIMIT 200

#Code 8
SELECT SUM(sales_proportion)
FROM (SELECT producer_id, 
        COUNT(producer_id) AS sales_count, 
        COUNT(producer_id)/1599828*100 AS sales_proportion
	FROM sales_data
	GROUP BY producer_id
	ORDER BY sales_count DESC
    LIMIT 100) as sales   # change LIMIT to get the SUMS

#Code 9
SELECT DISTINCT(purchase_origin), 
		COUNT(DISTINCT(product_id)) AS count_dist, 
    COUNT(DISTINCT(product_id))/17883 AS dist_prop, 
		COUNT(purchase_origin) AS count_sales,
    COUNT(purchase_origin)/1599828 AS sales_prop
FROM sales_data
GROUP BY purchase_origin
ORDER BY count_sales DESC

#Code 10
SELECT AVG(sales.count_sales),
	    STD(sales.count_sales)
FROM (SELECT MONTH(purchase_date) AS month,
		   COUNT(purchase_id) AS count_sales
		FROM sales_data
		GROUP BY MONTH(purchase_date)) AS sales

#Code 11
SELECT MONTH(purchase_date) AS month,
		COUNT(purchase_id) AS count_sales,
       SUM(CASE WHEN purchase_device = 'eReaders' THEN 1 ELSE 0 END) AS ereader_sales,
       SUM(CASE WHEN purchase_device = 'Desktop' THEN 1 ELSE 0 END) AS desktop_sales,
       SUM(CASE WHEN purchase_device = 'Smart TV' THEN 1 ELSE 0 END) AS smarttv_sales
FROM sales_data
GROUP BY MONTH(purchase_date) 
ORDER BY MONTH(purchase_date);

#Code 12
SELECT MONTH(purchase_date) AS month,
		COUNT(purchase_id) AS count_sales,
       SUM(CASE WHEN product_niche = 'Negotiation' THEN 1 ELSE 0 END) AS negotiation_sales,
       SUM(CASE WHEN product_niche = 'Anxiety management' THEN 1 ELSE 0 END) AS anxiety_sales,
       SUM(CASE WHEN product_niche = 'Personal finance' THEN 1 ELSE 0 END) AS pfinance_sales,
       SUM(CASE WHEN product_niche = 'Presentation skills' THEN 1 ELSE 0 END) AS presentations_sales
FROM sales_data
GROUP BY MONTH(purchase_date) 
ORDER BY MONTH(purchase_date);
```
